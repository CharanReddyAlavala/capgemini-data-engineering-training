```python
%sql
CREATE TABLE emp1 (
emp_id INT,
full_name STRING,
email STRING,
department STRING,
city STRING,
salary STRING,
remarks STRING
);
```

```python
%sql
INSERT INTO emp1 VALUES
(1, 'Karthik Kondpak', 'karthik.k@gmail.com', 'Data Engineering', 'Hyderabad', '75000', ' Top performer '),
(2, 'Veena Reddy', 'veena_r@company.com', 'Analytics', 'Bangalore', '65000', 'Excellent communication'),
(3, 'Ravi kumar', 'ravi.kumar@org.in', 'Data Science', 'Chennai', '85000', 'Needs improvement'),
(4, 'Anil', 'anil@abc.com', 'DEVOPS', 'Pune', '70000', NULL),
(5, ' Suresh ', 'suresh@xyz.com', 'data engineering', ' hyderabad ', '60000', ' ');
```

```python
%sql
select * from emp1;
```

LENGTH / CHAR_LENGTH

```python
%sql
SELECT full_name, LENGTH(full_name) FROM emp1;
```

UPPER / LOWER

```python
%sql
SELECT UPPER(department), LOWER(city) FROM emp1;
```

TRIM / LTRIM / RTRIM

```python
%sql
SELECT 
TRIM(full_name),
LTRIM(full_name),
RTRIM(full_name)
FROM emp1;
```

CONCAT

```python
%sql
SELECT CONCAT(full_name, ' - ', department) AS emp_details
FROM emp1;
```

CONCAT_WS

```python
%sql
SELECT CONCAT_WS(' | ', emp_id, full_name, city)
FROM emp1;
```

SUBSTRING / SUBSTR

```python
%sql
SELECT SUBSTRING(email, 1, 7) FROM emp1;
```

LEFT / RIGHT

```python
%sql
SELECT LEFT(full_name, 4), RIGHT(city, 3)
FROM emp1;
```

INSTR

```python
%sql
SELECT email, INSTR(email, '@') FROM emp1;
```

LOCATE

```python
%sql
SELECT email, LOCATE('.', email) FROM emp1;
```

REPLACE

```python
%sql
SELECT REPLACE(department, 'Data', 'Big Data')
FROM emp1;
```

REVERSE

```python
%sql
SELECT full_name, REVERSE(full_name)
FROM emp1;
```

LPAD / RPAD

```python
%sql
SELECT LPAD(emp_id, 5, '0') FROM emp1;
```

```python
%sql
SELECT RPAD(city, 15, '*') FROM emp1;
```

TRIM + REPLACE


```python
%sql
SELECT TRIM(REPLACE(city, ' ', '')) FROM emp1;
```

IFNULL → Databricks uses COALESCE

```python
%sql
SELECT full_name, COALESCE(remarks, 'No remarks')
FROM emp1;
```

COALESCE

```python
%sql
SELECT full_name, COALESCE(remarks, 'N/A')
FROM emp1;
```

ARRAY_CONTAINS

```python
%sql
SELECT ARRAY_CONTAINS(SPLIT('Data,Analytics,AI', ','), 'Analytics');
```
