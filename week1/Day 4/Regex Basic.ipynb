Q1: Numbers at the beginning

```python
%sql
SELECT mixed_value,
       regexp_extract(mixed_value, '^[0-9]+', 0) AS start_numbers
FROM regex_practice;
```

Q2: Numbers at the end

```python
%sql
SELECT mixed_value,
       regexp_extract(mixed_value, '[0-9]+$', 0) AS end_numbers
FROM regex_practice;
```

Q3: First character

```python
%sql
SELECT mixed_value,
       regexp_extract(mixed_value, '^.', 0) AS first_char
FROM regex_practice;
```

Q4: Last character

```python
%sql
SELECT mixed_value,
       regexp_extract(mixed_value, '.$', 0) AS last_char
FROM regex_practice;
```

Q5: Exactly 2 digits anywhere

```python
%sql
SELECT mixed_value,
       regexp_extract(mixed_value, '[0-9]{2}', 0) AS two_digits
FROM regex_practice;
```

Q6: Exactly 1 digit

```python
%sql
SELECT mixed_value,
       regexp_extract(mixed_value, '[0-9]', 0) AS one_digit
FROM regex_practice;
```

Q7: Country code from phone

```python
%sql
SELECT phone,
       regexp_extract(phone, '\\+?(\\d+)', 1) AS country_code
FROM regex_practice;
```

Q8: Numbers between alphabets

```python
%sql
SELECT mixed_value,
       regexp_extract(mixed_value, '[a-zA-Z]+([0-9]+)[a-zA-Z]+', 1) AS middle_numbers
FROM regex_practice;
```

Q9: Email before @

```python
%sql
SELECT email,
       regexp_extract(email, '^[^@]+', 0) AS username
FROM regex_practice;
```

Q10: After @ (domain + extension)

```python
%sql
SELECT email,
       regexp_extract(email, '@(.+)', 1) AS domain_full
FROM regex_practice;
```

Q11: Domain without @

```python
%sql
SELECT email,
       regexp_extract(email, '@([^@]+)', 1) AS domain
FROM regex_practice;
```

Q12: After last dot

```python
%sql
SELECT email,
       regexp_extract(email, '\\.([^.]+)$', 1) AS extension
FROM regex_practice;
```

Q13: Only alphabets

```python
%sql
SELECT mixed_value,
       regexp_extract(mixed_value, '[a-zA-Z]+', 0) AS alphabets
FROM regex_practice;
```

Q14: Only numbers

```python
%sql
SELECT mixed_value,
       regexp_extract(mixed_value, '[0-9]+', 0) AS numbers
FROM regex_practice;
```

Q15: First 3 characters

```python
%sql
SELECT full_text,
       regexp_extract(full_text, '^.{3}', 0) AS first_three
FROM regex_practice;
```

Q16: Last 2 characters


```python
%sql
SELECT full_text,
       regexp_extract(full_text, '.{2}$', 0) AS last_two
FROM regex_practice;
```

Q17: Employee number (between text & underscore)

```python
%sql
SELECT full_text,
       regexp_extract(full_text, '[A-Z]+([0-9]+)_', 1) AS emp_number
FROM regex_practice;
```

Q18: Country code at end


```python
%sql
SELECT full_text,
       regexp_extract(full_text, '([0-9]+)$', 1) AS country_code
FROM regex_practice;
```

Q19: Text between underscores

```python
%sql
SELECT full_text,
       regexp_extract(full_text, '_([A-Z]+)_', 1) AS middle_text
FROM regex_practice;
```

Q20: Digits after +

```python
%sql
SELECT phone,
       regexp_extract(phone, '\\+(\\d+)', 1) AS country_code
FROM regex_practice;
```
