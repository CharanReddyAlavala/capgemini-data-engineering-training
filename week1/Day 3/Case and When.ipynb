```python
%sql
SELECT *,CASE WHEN experience >= '8' AND performance_rating = 'A' THEN '20%'
WHEN experience >= '5' AND performance_rating = 'B' THEN '15%'
WHEN  performance_rating = 'C' Then 'No'
ELSE '10%' END AS Hike FROM Emp;
```

```python
%sql
SELECT *, CASE WHEN department = 'Finance' THEN 
CASE WHEN performance_rating = 'A' THEN "20%" 
WHEN performance_rating = 'B' THEN '15%'
WHEN performance_rating = 'C' THEN '5%'
END 
WHEN department = "Engineering" THEN 
CASE WHEN performance_rating = 'A' THEN "18%"
WHEN performance_rating = 'B' THEN '12%' 
WHEN performance_rating = 'C' THEN '3%'
END 
ELSE '10%' END AS Bonus FROM Emp;
```

```python
%sql
SELECT *, CASE
WHEN salary > '80000' AND performance_rating = 'A' THEN 'High'
WHEN salary > '50000' AND performance_rating = 'B' THEN 'Mid'
WHEN salary < '20000' OR  performance_rating = 'C' Then 'Low'
END AS Performer from Emp;
```

```python
%sql
SELECT *,CASE 
WHEN department = 'HR' THEN 
CASE 
WHEN experience < '5' THEN 'High'
WHEN experience > '5' THEN 'Low'
END
WHEN department IN ('Engineering', 'Finance') THEN 
CASE 
WHEN experience > '8' THEN 'Low'
WHEN experience < '8' THEN 'Mid'
END
ELSE 'Mid' END AS Risk FROM Emp;
```

```python
%sql
SELECT *,CASE
WHEN performance_rating = 'A' THEN
CASE
WHEN salary > '80000' and experience > '5' THEN '25%'
WHEN salary > '50000' and salary < '80000' THEN '15%'
ELSE '20%'
END
WHEN performance_rating = 'B' THEN
CASE
WHEN experience > '5' THEN '12%'
ELSE '10%'
END
WHEN performance_rating = 'C' THEN 'NO'
END AS Hike from Emp;
```

```python
%sql
SELECT *,CASE
WHEN department = 'Finance' THEN 
CASE 
WHEN performance_rating = 'A' and experience > '10' THEN '25%'
ELSE '20%'
END
WHEN department = 'HR' THEN
CASE 
WHEN performance_rating = 'B' or experience > '5' THEN '10%'
ELSE '10%'
END
ELSE '8%' 
END AS Bonus from emp;
```

```python
%sql
SELECT *, CASE 
WHEN salary > '70000' THEN 
CASE
WHEN performance_rating = 'A' and experience = '8' THEN 'Top'
WHEN experience < '8' THEN 'Mid'
END
WHEN salary between '50000' and '70000' and performance_rating ='A' THEN 'Rising Star'
WHEN salary between '50000' and '70000' THEN 'Average'
WHEN salary < '50000' THEN 'Low'
END AS Performer from emp;
```

```python
%sql
SELECT *, CASE
WHEN SALARY > 90000 THEN 
CASE 
WHEN experience > 10 THEN '35%'
ELSE '30%'
END
WHEN SALARY BETWEEN 60000 AND 90000 THEN
CASE 
WHEN experience > 5 THEN '25%'
ELSE '20%'
END 
WHEN SALARY < 60000 THEN '15%'
END AS Tax from emp;
```
